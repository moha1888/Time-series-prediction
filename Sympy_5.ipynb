```python
import sympy as sp
%matplotlib inline
```

```python
sp.__version__
```

```python
x = sp.symbols('x')
```

```python
f = x**2 - 2*x +1
```

```python
sp.pprint(f)
```

```python
print( 'Evaluate function at x = 2')
print(f.subs(x,2))
```

plotting the functions

plot the function called f

```python
p1 = plot(f, show=False)
p1.show()
```
